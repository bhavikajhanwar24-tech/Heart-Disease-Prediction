```python
import pickle

with open("heart_disease_model.pkl", "rb") as f:
    model = pickle.load(f)
```

```python
import pandas as pd

def predict_heart_disease(input_data, threshold=0.4):
    df = pd.DataFrame([input_data])
    
    prob = model.predict_proba(df)[0][1]
    prediction = 1 if prob >= threshold else 0
    
    return {
        "probability": round(prob, 3),
        "prediction": prediction
    }
```

```python
patient = {
    'age': 57,
    'sex': 'female',
    'chest_pain_type': 'atypical angina',
    'resting_blood_pressure': 130,
    'cholesterol': 236,
    'fasting_blood_sugar': 'lower than 120mg/ml',
    'resting_electrocardiogram': 'normal',
    'max_heart_rate_achieved': 174,
    'exercise_induced_angina': 'no',
    'st_depression': 0,
    'st_slope': 'flat',
    'num_major_vessels': 1,
    'thalassemia': 'normal'
}

result = predict_heart_disease(patient)

print("Probability:", result["probability"])
print("Prediction:", result["prediction"])
```
